```python
def f(x):
    if isinstance(x,list):
       return [val **6 + 10 * x**3 - 1000 for val in x]
    return x**6 + 10 * x**3 - 1000
```

```python
import sympy as sym 

x = sym.Symbol('x')

sym.diff(x**6 + 10 *x**3 - 1000 )
```

```python
def f_diff(x):
    if isinstance(x,list):
        return [val **5 * 6 + 30 * val * 2 for val in x]
    return 6*x ** 5 + 30*x**2
```

```python
#6 Чему равна производная при x = 10
f_diff(10)
```

```python
#7 Покажите на графике f(x) & f_diff(x)
```

```python
x_diff = np.arange(-10,11)
y_diff = f_diff(x=x_diff)

plt.plot(x_points, y_points, marker='o', linestyle='--', color='g', label='Before Diff')
plt.plot(x_diff,y_diff, marker ='o', linestyle='--', color='r', label= 'After Diff')

plt.legend()
plt.show()
```
