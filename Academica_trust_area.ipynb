```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

#  1 -  
mean_brochure = 20
std_brochure = 2
threshold = 18

# ,      18 
probability_brochure = stats.norm.cdf(threshold, loc=mean_brochure, scale=std_brochure)

#  2 -   
mean_height = 175
std_height = 6
confidence_level = 0.95
sample_size = 30  #  

#    (SEM)
sem_height = std_height / np.sqrt(sample_size)

#  
ci_height = stats.norm.interval(confidence_level, loc=mean_height, scale=sem_height)

#    
data_height = np.random.normal(mean_height, std_height, 1000)

#  3 -  
mean_score = 75
std_score = 10
sample_size_scores = 100

#  
scores = np.random.normal(mean_score, std_score, sample_size_scores)
point_estimate = np.mean(scores)

#  KDE-   2
plt.figure(figsize=(10, 6))
sns.kdeplot(data_height, fill=True, label='  ', color='blue')
plt.axvline(ci_height[0], color='red', linestyle='--', label=f' : {ci_height[0]:.2f} ')
plt.axvline(ci_height[1], color='green', linestyle='--', label=f' : {ci_height[1]:.2f} ')
plt.title('95%     ')
plt.xlabel(' ()')
plt.ylabel(' ')
plt.legend()
plt.show()

probability_brochure, ci_height, point_estimate
```
