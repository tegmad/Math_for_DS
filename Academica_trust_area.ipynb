```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Задача 1 - Вероятность распределения
mean_brochure = 20
std_brochure = 2
threshold = 18

# Вероятность, что случайная брошюра имеет менее 18 страниц
probability_brochure = stats.norm.cdf(threshold, loc=mean_brochure, scale=std_brochure)

# Задача 2 - Оценка доверительного интервала
mean_height = 175
std_height = 6
confidence_level = 0.95
sample_size = 30  # предполагаемая выборка

# Стандартная ошибка среднего (SEM)
sem_height = std_height / np.sqrt(sample_size)

# Доверительный интервал
ci_height = stats.norm.interval(confidence_level, loc=mean_height, scale=sem_height)

# Генерация данных для визуализации
data_height = np.random.normal(mean_height, std_height, 1000)

# Задача 3 - Точечная оценка
mean_score = 75
std_score = 10
sample_size_scores = 100

# Генерация выборки
scores = np.random.normal(mean_score, std_score, sample_size_scores)
point_estimate = np.mean(scores)

# Визуализация KDE-графика для Задачи 2
plt.figure(figsize=(10, 6))
sns.kdeplot(data_height, fill=True, label='Распределение роста студентов', color='blue')
plt.axvline(ci_height[0], color='red', linestyle='--', label=f'Нижняя граница: {ci_height[0]:.2f} см')
plt.axvline(ci_height[1], color='green', linestyle='--', label=f'Верхняя граница: {ci_height[1]:.2f} см')
plt.title('95% доверительный интервал для роста студентов')
plt.xlabel('Рост (см)')
plt.ylabel('Плотность вероятности')
plt.legend()
plt.show()

probability_brochure, ci_height, point_estimate
```
