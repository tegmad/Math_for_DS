```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns

mu1, sigma1 = 30, 1
x1 = 29

z1 = (x1 - mu1) / sigma1
p1 = norm.cdf(z1)

print(f"Вероятность того, что порция будет менее 29 мл: {p1:.4f}")

alpha1 = 0.05
if p1 < alpha1:
    print("Реальная практика НЕ соответствует стандартам кафе.")
else:
    print("Реальная практика соответствует стандартам кафе.")

mu2, sigma2 = 18, 4
x2 = 22

z2 = (x2 - mu2) / sigma2
p2 = 1 - norm.cdf(z2)

print(f"Вероятность того, что снижение уровня холестерина будет 22 мг/дл или больше: {p2:.4f}")

alpha2 = 0.01
if p2 < alpha2:
    print("Такое снижение холестерина НЕ является случайностью.")
else:
    print("Такое снижение холестерина является случайностью.")

np.random.seed(42)

norm_data = np.random.normal(50, 10, 1000)
uniform_data = np.random.uniform(10, 60, 1000)
exp_data = np.random.exponential(10, 1000)

print(f"Среднее значение нормального распределения: {np.mean(norm_data):.2f}")
print(f"Среднее значение равномерного распределения: {np.mean(uniform_data):.2f}")
print(f"Среднее значение экспоненциального распределения: {np.mean(exp_data):.2f}")

sample = np.random.choice(norm_data, 300, replace=False)
print(f"Среднее значение выборки из нормального распределения: {np.mean(sample):.2f}")

sns.histplot(norm_data, kde=True, bins=30, color='blue', label='Нормальное распределение')
sns.histplot(uniform_data, kde=True, bins=30, color='green', label='Равномерное распределение')
sns.histplot(exp_data, kde=True, bins=30, color='red', label='Экспоненциальное распределение')
plt.legend()
plt.title("Гистограммы распределений")
plt.show()

import statsmodels.api as sm
sm.qqplot(norm_data, line='s')
plt.title("QQ Plot для нормального распределения")
plt.show()

mu4, sigma4, n4 = 30, 10, 50
sample_mean4 = np.random.normal(mu4, sigma4 / np.sqrt(n4), 1000)

z_lower = (30 - 1.96 * sigma4 / np.sqrt(n4) - mu4) / (sigma4 / np.sqrt(n4))
z_upper = (30 + 1.96 * sigma4 / np.sqrt(n4) - mu4) / (sigma4 / np.sqrt(n4))

p4 = norm.cdf(z_upper) - norm.cdf(z_lower)
print(f"Вероятность, что среднее время в пути близко к 30 минутам: {p4:.4f}")
```
