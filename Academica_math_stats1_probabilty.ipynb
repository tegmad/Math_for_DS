```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
np.random.seed(42)

sample = np.random.randint(1,7,1000)
```

```python
plt.figure(figsize=(8,6))
plt.hist(sample, bins=np.arange(1,7),rwidth=0.7, density=True)
plt.xlabel('Numbers')
plt.ylabel('Probability')
plt.show()
```

```python

import statsmodels.api as sm

np.random.seed(42)

mean, std_dev, size = 50, 15, 10000
original_data = np.random.normal(mean, std_dev, size)

outliers = np.array([200] * 10)
modified_data = np.concatenate([original_data, outliers])

original_mean, original_std = np.mean(original_data), np.std(original_data)
modified_mean, modified_std = np.mean(modified_data), np.std(modified_data)

plt.figure(figsize=(10, 6))
plt.hist(modified_data, bins=50, alpha=0.7, label='Modified Data')
plt.axvline(modified_mean, color='r', linestyle='--', label=f'Modified Mean: {modified_mean:.2f}')
plt.axvline(modified_mean + modified_std, color='g', linestyle='--', label=f'Modified Std Dev: {modified_std:.2f}')
plt.axvline(modified_mean - modified_std, color='g', linestyle='--')
plt.title('Effect of Outliers on Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
cityA_data = np.random.normal(50000, 15000, 1000)
cityB_data = np.random.normal(55000, 20000, 1000)

plt.figure(figsize=(10, 6))
sns.kdeplot(cityA_data, shade=True, label='City A', color='blue')
sns.kdeplot(cityB_data, shade=True, label='City B', color='orange')
plt.axvline(np.mean(cityA_data), color='blue', linestyle='--', label=f'City A Mean: {np.mean(cityA_data):.2f}')
plt.axvline(np.mean(cityB_data), color='orange', linestyle='--', label=f'City B Mean: {np.mean(cityB_data):.2f}')
plt.title('Income Distribution in City A and City B')
plt.xlabel('Income')
plt.ylabel('Density')
plt.legend()
plt.show()

population = np.random.exponential(scale=10, size=1000)
sample = np.random.choice(population, size=100, replace=False)

plt.figure(figsize=(10, 6))
sns.kdeplot(sample, shade=True, label='Sample KDE')
plt.hist(sample, bins=20, alpha=0.7, label='Sample Histogram', color='orange', density=True)
plt.title('Sample Distribution of Service Time')
plt.xlabel('Service Time')
plt.ylabel('Density')
plt.legend()
plt.show()

sm.qqplot(sample, line='r')
plt.title('Q-Q Plot for Normality Check')
plt.show()
```
