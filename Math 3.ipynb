```python
#1. Создайте функцию f(x) = 3 * x^2 + 2 * x - 10
```

```python
def f(x):
    if isinstance(x,list):
        return [3 * (val**2) + 2*x -10 for val in x]
    return 3 *(x**2) + 2*x - 10
```

```python
#11. Посчитайте интеграл функции f(x) между 0 и 5. Также отобразите на графике площадь, которую мы посчитали 
```

```python
from scipy.integrate import quad

# Задаем функцию f(x) для интегрирования
def f(x):
    return 3 * x**2 + 2 * x - 10

# Вычисление определенного интеграла от 0 до 5
integral_result, _ = quad(f, 0, 5)

# Построение графика функции и закрашенной области под кривой
x_vals = np.linspace(-1, 6, 100)
y_vals = f(x_vals)

# Пределы интегрирования для закрашенной области
x_fill = np.linspace(0, 5, 100)
y_fill = f(x_fill)

# Построение графиков
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r'$f(x) = 3x^2 + 2x - 10$', color='blue')
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.4, label=f'Area (Integral) ≈ {integral_result:.2f}')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Graph of f(x) with Shaded Integral Area from 0 to 5")
plt.legend()
plt.grid(True)
plt.show()

integral_result
```
